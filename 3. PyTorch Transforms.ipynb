# Transforms

The data generally does not always come in its final processed form which is required for training machine learning algorithms. To solve this, we use transforms to manipulate the data amd make it suitable for training.

```python
import torch
from torchvision import datasets
from torchvision.transforms import ToTensor, Lambda

ds = datasets.FashionMNIST(root= "data", train= True, download= True, transform= ToTensor(), target_transform= Lambda(lambda y: torch.zeros(10, dtype= torch.float). scatter_(0, torch.tensor(y), value= 1)))
```

The TorchVision datasets have two parameters:
* transform: to modify the features
* target_transform: to modify the labels - that accept callables containing the transformation logic

### Lambda Transforms

```python
target_transform = Lambda(lambda y: torch.zeros(10, dtype= torch.float).scatter_(dim=0, index= torch.tensor(y), value= 1))
```
